```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import gaussian
from numpy.fft import fft2, ifft2
```

```python
image = cv.imread('source2.jpg')
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
gamma = 0.4
gamma_corrected = np.power(image/255.0, gamma) * 255.0
gamma_corrected = np.uint8(gamma_corrected)
plt.imshow(image)
plt.imshow(gamma_corrected)
```

```python
mean_value = np.mean(gamma_corrected)
adjusted_image = gamma_corrected - mean_value

contrast_factor = 3.5
contrast_adjusted = np.clip(adjusted_image * contrast_factor + mean_value, 0, 255)

contrast_adjusted = np.uint8(contrast_adjusted)
plt.imshow(contrast_adjusted)
```
