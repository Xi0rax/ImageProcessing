```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread('inscription.jpg', cv.IMREAD_GRAYSCALE)
plt.figure(figsize=(14, 10))
plt.imshow(image, cmap='gray')
```

```python
thresholds = [122, 128]

prev_image = image

for tr in thresholds:
    ret, prev_image = cv.threshold(prev_image, tr, 255, cv.THRESH_BINARY)
plt.figure(figsize=(14, 10))
plt.imshow(prev_image, cmap='gray')
```

```python
cv.imwrite('res.jpg', prev_image)
```
